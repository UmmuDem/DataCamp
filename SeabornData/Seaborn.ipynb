## Import Matplotlib and Seaborn

```python
import matplotlib.pyplot as plt 
import seaborn as sns
import pandas as pd
```

```python
countries = pd.read_csv('countries-of-the-world.csv')
```

```python
countries.head()
```

```python
gdp = list(countries['GDP ($ per capita)'])
```

```python
phones = list(countries['Phones (per 1000)'])
```

```python
region = list(countries['Region'])
```

**Making a scatter plot with lists**

```python
# Create scatter plot with GDP on the x-axis and number of phones on the y-axis
sns.scatterplot(x=gdp, y=phones)

# Show plot
plt.show()
```

**Making a count plot with a list**

```python

# Create count plot with region on the y-axis
sns.countplot(y=region)

# Show plot
plt.show()
```

## Using pandas with Seaborn

```python

# Create a DataFrame from csv file
df = pd.read_csv('young-people-survey-responses.csv')

# Create a count plot with "Spiders" on the x-axis
sns.countplot(x='Spiders', data = df)

# Display the plot
plt.show()
```

**Hue and scatter plots**

```python
student_data = pd.read_csv('student-alcohol-consumption.csv')
# Change the legend order in the scatter plot
sns.scatterplot(x="absences", y="G3", 
                data=student_data, 
                hue="location", hue_order = ['Rural','Urban'])

# Show plot
plt.show()
```

**Hue and count plots**

```python
# Create a dictionary mapping subgroup values to colors
palette_colors = {'Rural': "green", 'Urban': "blue"}

# Create a count plot of school with location subgroups
sns.countplot(x='school', data=student_data, hue='location', palette = palette_colors)


# Display plot
plt.show()
```

**Creating subplots with col and row**

```python
sns.relplot(x="absences", y="G3", 
            data=student_data,
            kind="scatter", 
            row="study_time")  # or we can use col = 'study_time'

# Show plot
plt.show()
```

**Creating two-factor subplots**

```python
# Adjust further to add subplots based on family support
sns.relplot(x="G1", y="G3", 
            data=student_data,
            kind="scatter", 
            col="schoolsup",
            col_order=["yes", "no"],
            row='famsup', row_order=['yes','no'])

# Show plot
plt.show()
```

**Changing the size of scatter plot points**

```python
mpg = pd.read_csv('mpg.csv')
# Create scatter plot of horsepower vs. mpg
sns.relplot(x="horsepower", y="mpg", 
            data=mpg, kind="scatter", 
            hue = 'cylinders',
            size="cylinders")

# Show plot
plt.show()
```

**Changing the style of scatter plot points**

```python
# Create a scatter plot of acceleration vs. mpg
sns.relplot(x='acceleration', y='mpg', data=mpg, kind='scatter',
style='origin', hue='origin')



# Show plot
plt.show()
```

**Visualizing standard deviation with line plots**

```python
# Make the shaded area show the standard deviation
sns.relplot(x="model_year", y="mpg",
            data=mpg, kind="line", ci='sd')   # ci:confidence interval, here it is set as standard deviation, by default it is mean.

# Show plot
plt.show()
```

**Plotting subgroups in line plots**

```python

# Create line plot of model year vs. horsepower

sns.relplot(x='model_year',y='horsepower', data=mpg, kind='line',ci=None)


# Show plot
plt.show()
```

```python
# Add markers and make each line have the same style
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, style="origin", 
            hue="origin")

# Show plot
plt.show()
```

```python
# Add markers and make each line have the same style
sns.relplot(x="model_year", y="horsepower", 
            data=mpg, kind="line", 
            ci=None, style="origin", 
            hue="origin",markers=True, dashes=False)

# Show plot
plt.show()
```

## Count plots and bar plots

**Count plots**

```python
survey_data = pd.read_csv('young-people-survey-responses.csv')
# Separate into column subplots based on age category
sns.catplot(y="Internet usage", data=survey_data,
            kind="count",
            col='Gender')

# Show plot
plt.show()
```

**Bar plots with percentages**

```python
# Create a bar plot of interest in math, separated by gender
sns.catplot(x='Gender',y='Mathematics', data=survey_data,kind='bar')


# Show plot
plt.show()
```

**Customizing bar plots**

```python
# List of categories from lowest to highest
category_order = ["<2 hours", 
                  "2 to 5 hours", 
                  "5 to 10 hours", 
                  ">10 hours"]

sns.catplot(x="study_time", y="G3",
            data=student_data,
            kind="bar",
            order=category_order)

# Show plot
plt.show()
```

```python
# List of categories from lowest to highest
category_order = ["<2 hours", 
                  "2 to 5 hours", 
                  "5 to 10 hours", 
                  ">10 hours"]

# Turn off the confidence intervals
sns.catplot(x="study_time", y="G3",
            data=student_data,
            kind="bar",
            order=category_order, ci=None)

# Show plot
plt.show()
```

### Boxplots

**Create and interpret a box plot**

```python
# List of categories from lowest to highest
category_order = ["<2 hours", 
                  "2 to 5 hours", 
                  "5 to 10 hours", 
                  ">10 hours"]

sns.catplot(x="study_time", y="G3",
            data=student_data,
            kind="box",
            order=category_order)

# Show plot
plt.show()
```

**Omitting outliers**

```python
# Create a box plot with subgroups and omit the outliers
sns.catplot(x='internet',y='G3',data=student_data, kind='box',hue='location', sym='')  # sym parameter omits the outliers.

# Show plot
plt.show()
```

**Adjusting the whiskers**

```python
# Set the whiskers to 0.5 * IQR
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box", whis=0.5)

# Show plot
plt.show()
```

```python
# Extend the whiskers to the 5th and 95th percentile
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            whis=[5,95])

# Show plot
plt.show()
```

```python
# Extend the whiskers to the 5th and 95th percentile
sns.catplot(x="romantic", y="G3",
            data=student_data,
            kind="box",
            whis=[0,100])

# Show plot
plt.show()
```

### Pointplots

**Customizing point plots**

```python
sns.catplot(x="famrel", y="absences",data=student_data,
            kind="point",
            capsize=0.2)
            
# Show plot
plt.show()
```

```python
# Remove the lines joining the points
sns.catplot(x="famrel", y="absences",
			data=student_data,
            kind="point",
            capsize=0.2,join=False)
            
# Show plot
plt.show()
```

**Point plots with subgroups**

```python
# Create a point plot that uses color to create subgroups
sns.catplot(x='romantic',y='absences',data=student_data,kind='point',hue='school')

# Show plot
plt.show()
```

```python
# Turn off the confidence intervals for this plot
sns.catplot(x="romantic", y="absences",
			data=student_data,
            kind="point",
            hue="school", ci=None)

# Show plot
plt.show()
```

```python
# Import median function from numpy
from numpy import median

# Plot the median number of absences instead of the mean
sns.catplot(x="romantic", y="absences",
			data=student_data,
            kind="point",
            hue="school",
            ci=None, estimator=median)

# Show plot
plt.show()
```

### Adding Titles and Labels

**FacetGrids vs. AxesSubplots**

```python
# Create scatter plot
g = sns.relplot(x="weight", 
                y="horsepower", 
                data=mpg,
                kind="scatter")

# Identify plot type
type_of_g = type(g)

# Print type
print(type_of_g)
```

**Adding a title to a FacetGrid object**

```python
# Create scatter plot
g = sns.relplot(x="weight", 
                y="horsepower", 
                data=mpg,
                kind="scatter")

# Add a title "Car Weight vs. Horsepower"
g.fig.suptitle("Car Weight vs. Horsepower",y=1.05)

# Show plot
plt.show()
```

```python
# Create line plot
g = sns.lineplot(x="model_year", y="mpg", 
                 data=mpg,
                 hue="origin",ci=None)

# Add a title "Average MPG Over Time"
g.set_title("Average MPG Over Time")

# Show plot
plt.show()
```

```python
# Create line plot
g = sns.lineplot(x="model_year", y="mpg", 
                 data=mpg,
                 hue="origin")

# Add a title "Average MPG Over Time"
g.set_title("Average MPG Over Time")

# Add x-axis and y-axis labels
g.set(xlabel="Car Model Year",ylabel="Average MPG")



# Show plot
plt.show()
```
